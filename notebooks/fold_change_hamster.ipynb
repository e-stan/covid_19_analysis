```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
import matplotlib
matplotlib.rcParams['figure.dpi'] = 500
from brokenaxes import brokenaxes
```

```python
filename = "../data/long_to_plot_fc_hamster_STAT.csv"
```

```python
stats = ["COV","Inf"]
color_code = {"LPC":(.502,0,.502),"PC":(0,0,1),"Cer_NS":(.1,.1,.1),"_LPE":(.7021,0,0),"Serine":"green"}
for stat in stats:
    legend_entries = []
    dat = pd.read_csv(filename.replace("STAT",stat),index_col=0)
    xvals = [x for x in range(len(dat.columns.values))]
    fig,ax1 = plt.subplots()
    #ax2 = ax1.twinx()
   # bax = brokenaxes(ylims = ((0,2),(7,8.5)))
    for index,row in dat.iterrows():
        if any(index.startswith(k) for k in color_code):  
            for k in color_code:
                if index.startswith(k):
                    if type(color_code[k]) == type((1,1)):
                        color = tuple(color_code[k])
                        if k not in legend_entries:
                            ax1.plot([0,0],[0,0],color=color,label=k.replace("_",""))
                            legend_entries.append(k)
                        color = tuple(np.array(color) * (.3*np.random.random() + .70))
                    else:
                        color = str(color_code[k])
                        if k not in legend_entries:
                            ax1.plot([0,0],[0,0],color=color,label=k.replace("_",""))
                            legend_entries.append(k)
            ax1.plot(xvals,row.values,color=color,alpha=0.5)
            #plt.scatter(xvals,row.values,color=color)
            
        else:
            if "Serine" in index:
                ax1.plot(xvals,row.values,label=index,alpha=0.5)
            else:
                ax1.plot(xvals,row.values,label=index,alpha=0.5)

            #plt.scatter(xvals,row.values)

    plt.title(stat)
    ax1.set_ylim((0,2))
    #bax.set_xticks(xvals)
    #bax.set_xticklabels(labels=dat.columns.values)
    plt.xticks(xvals,labels=dat.columns.values,fontsize=8)
    plt.yticks(fontsize=8)
    #plt.ylabel("FC")
    plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))
    #plt.ylim((0,2))
    ax1.plot([min(xvals),max(xvals)],[1.0,1.0],color="grey",alpha=0.5,ls="--")
```

```python
stats = ["COV","Inf"]
color_code = {"LPC":(.502,0,.502),"PC":(0,0,1),"Cer_NS":(.1,.1,.1),"_LPE":(.7021,0,0),"Serine":"green"}
for stat in stats:
    legend_entries = []
    dat = pd.read_csv(filename.replace("STAT",stat),index_col=0)
    xvals = [x for x in range(len(dat.columns.values))]
    fig,ax1 = plt.subplots()
    ax2 = ax1.twinx()
   # bax = brokenaxes(ylims = ((0,2),(7,8.5)))
    for index,row in dat.iterrows():
        if any(index.startswith(k) for k in color_code):  
            for k in color_code:
                if index.startswith(k):
                    if type(color_code[k]) == type((1,1)):
                        color = tuple(color_code[k])
                        if k not in legend_entries:
                            ax1.plot([0,0],[0,0],color=color,label=k.replace("_",""))
                            legend_entries.append(k)
                        color = tuple(np.array(color) * (.3*np.random.random() + .70))
                    else:
                        color = str(color_code[k])
                        if k not in legend_entries:
                            ax1.plot([0,0],[0,0],color=color,label=k.replace("_",""))
                            legend_entries.append(k)
            if "Serine" in index:
                ax2.plot(xvals,row.values,label=index,alpha=0.5,color=color)
            else:
                ax1.plot(xvals,row.values,label=index,alpha=0.5,color=color)
            
        else:
            if "Serine" in index:
                ax2.plot(xvals,row.values,label=index,alpha=0.5)
            else:
                ax1.plot(xvals,row.values,label=index,alpha=0.5)

            #plt.scatter(xvals,row.values)

    plt.title(stat)
    ax1.set_ylim((0,1.5))
    #bax.set_xticks(xvals)
    #bax.set_xticklabels(labels=dat.columns.values)
    plt.xticks(xvals,labels=dat.columns.values,fontsize=8)
    ax1.set_xticklabels(dat.columns.values,fontsize=8)
    ax2.tick_params(color="green",labelsize=8,labelcolor="green")
    ax1.tick_params(labelsize=8)

    plt.yticks(fontsize=8)
    #plt.ylabel("FC")
    #plt.gca().legend(loc='center left', bbox_to_anchor=(1, 0.5))
    #plt.ylim((0,2))
    ax1.plot([min(xvals),max(xvals)],[1.0,1.0],color="grey",alpha=0.5,ls="--")
    ax2.plot([min(xvals),max(xvals)],[1.0,1.0],color="green",alpha=0.2,ls="--")
    ax1.set_ylabel("FC")
    ax2.set_ylabel("FC (Serine)",color="green")
```
